```python
import json
from pathlib import Path
import pandas as pd

pilot_path = Path("/workspace/thesis/outputs_raw/pilot/pilot_run_01.jsonl")

rows = []
with pilot_path.open("r", encoding="utf-8") as f:
    for line in f:
        line = line.strip()
        if line:
            rows.append(json.loads(line))

len(rows)
```

```python
df = pd.DataFrame(rows)
df.head()
```

```python
df.columns.tolist()
```

```python
for c in df.columns:
    print(c, df[c].isna().sum())
```

```python
text_col = None
for candidate in ["prediction", "pred", "text", "answer", "outputs"]:
    if candidate in df.columns:
        text_col = candidate
        break

print("text_col =", text_col)
df[[c for c in df.columns if c in ["id", "study_id", "image", text_col]]].head(10)
```

```python
if text_col:
    df["text_len"] = df[text_col].astype(str).str.len()
    df["text_len"].describe()
```

```python
df.to_csv("/workspace/thesis/outputs_clean/analysis/pilot_run_01_preview.csv", index=False)
```
