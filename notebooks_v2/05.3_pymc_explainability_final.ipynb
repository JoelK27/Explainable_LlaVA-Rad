```python
"""
PIPELINE STAGE 5.3: Bayesian Latent Modeling & Downstream Regression for manually reviewed explainability scores (z_i)
------------------------------------------------------------------
Role in Thesis:
    Aggregates the observation-level overlap metrics into an uncertainty-aware, 
    latent explainability score (z_i) using PyMC. Performs downstream regression 
    to evaluate Research Question 2 (Hypothesis 2): Does higher explainability 
    correlate with lower hallucination burden?
    
Diagnostics Checked:
    - Zero Divergences
    - Maximum R-hat value (acceptable around 1.0)
    - Posterior Predictive Checks (PPC) on transformed precision/recall features
"""
```

```python
# =====================================================================
# ROBUST STANDARDIZATION
# =====================================================================
# Applies robust z-standardization (median/IQR) to precision and recall features 
# to keep extreme extraction outliers from distorting the PyMC likelihood space.

# =====================================================================
# BAYESIAN SCORE MODEL (LATENT EXPLAINABILITY E_i)
# =====================================================================
# Implements Eq. 3.12 from the thesis.
# Explainability is treated as a latent variable (E_i or z_i) that cannot be 
# observed directly. It is inferred probabilistically from imperfect indicators 
# (precision_obs and recall_obs).
# 
# - E: The unobserved Explainability score (latent space)
# - alpha: Feature-specific intercepts
# - sigma_shared: Residual variance not explained by E.

# =====================================================================
# DOWNSTREAM REGRESSION ANALYSIS (H2)
# =====================================================================
# Predicts hallucination burden (Hz) from the inferred explainability score (Sz).
# The parameter 'beta1' represents the substantive thesis outcome:
# A negative beta1 posterior indicates that higher structurally grounded 
# explainability corresponds to a lower rate of unsupported clinical findings.

# =====================================================================
# MCMC DIAGNOSTICS & POSTERIOR PREDICTIVE CHECKS
# =====================================================================
# Verifies model credibility per Gelman et al. / Vehtari et al.:
# - Extracts max R-hat (chain mixing parameter, acceptable around <= 1.02)
# - Validates lack of divergences.
# - Computes Posterior Predictive Checks (PPC) to ensure the transformed 
#   feature distributions are plausibly reproduced by the model.

import os
import sys

os.environ.setdefault("PYTENSOR_FLAGS", "cxx=,optimizer_excluding=fusion")
 
!{sys.executable} -m pip install -q pymc arviz pandas numpy matplotlib nutpie

from pathlib import Path
import pandas as pd
import numpy as np
import pymc as pm
import arviz as az
import matplotlib.pyplot as plt
from IPython.display import display

# ------------------------------------------------------------------
# Paths
# ------------------------------------------------------------------
PROJECT_ROOT = Path.cwd().parent

FEATURE_DATAPATH = PROJECT_ROOT / "outputs" / "radgraph_analysis_1500_subset" / "case_level_comparison.csv"
HALL_DATAPATH = PROJECT_ROOT / "outputs" / "manual_review_1500_subset" / "final_hallucination_outputs" / "final_hallucination_case_level.csv"

OUT = PROJECT_ROOT / "outputs" / "bayesian_explainability_1500_subset_manual_final_V2"
OUT.mkdir(parents=True, exist_ok=True)

print("Current working directory:", Path.cwd())
print("FEATURE_DATAPATH:", FEATURE_DATAPATH)
print("Resolved FEATURE_DATAPATH:", FEATURE_DATAPATH.resolve())
print("FEATURE_DATAPATH exists:", FEATURE_DATAPATH.exists())

print("HALL_DATAPATH:", HALL_DATAPATH)
print("Resolved HALL_DATAPATH:", HALL_DATAPATH.resolve())
print("HALL_DATAPATH exists:", HALL_DATAPATH.exists())

if not FEATURE_DATAPATH.exists():
    raise FileNotFoundError(f"Could not find {FEATURE_DATAPATH}")

if not HALL_DATAPATH.exists():
    raise FileNotFoundError(f"Could not find {HALL_DATAPATH}")

# ------------------------------------------------------------------
# Load data
# ------------------------------------------------------------------
df = pd.read_csv(FEATURE_DATAPATH)
hall_df = pd.read_csv(HALL_DATAPATH)

# ------------------------------------------------------------------
# Normalize columns
# ------------------------------------------------------------------
rename_map = {
    "caseid": "case_id",
    "generatedobservationcount": "generated_observation_count",
    "referenceobservationcount": "reference_observation_count",
    "sharedobservationcount": "shared_observation_count",
    "generatedonlycount": "generated_only_count",
    "referenceonlycount": "reference_only_count",
    "precisionobs": "precision_obs",
    "recallobs": "recall_obs",
    "f1obs": "f1_obs",
    "jaccardobs": "jaccard_obs",
    "sharedrateref": "shared_rate_ref",
    "generatedonlyrate": "generated_only_rate",
    "referenceonlyrate": "reference_only_rate",
    "mismatchtotal": "mismatch_total",
    "mismatchrateunion": "mismatch_rate_union",
    "highhallucinationbinary": "high_hallucination_binary",
}
df = df.rename(columns={c: rename_map.get(c, c) for c in df.columns})
hall_df = hall_df.rename(columns={c: rename_map.get(c, c) for c in hall_df.columns})

df["case_id"] = pd.to_numeric(df["case_id"], errors="coerce").astype("Int64")
hall_df["case_id"] = pd.to_numeric(hall_df["case_id"], errors="coerce").astype("Int64")

numeric_cols = [
    "generated_observation_count",
    "reference_observation_count",
    "shared_observation_count",
    "generated_only_count",
    "reference_only_count",
    "precision_obs",
    "recall_obs",
    "f1_obs",
    "jaccard_obs",
    "shared_rate_ref",
    "generated_only_rate",
    "reference_only_rate",
    "mismatch_total",
    "mismatch_rate_union",
]

for c in numeric_cols:
    if c in df.columns:
        df[c] = pd.to_numeric(df[c], errors="coerce")

hall_numeric_cols = [
    "n_reviewed",
    "n_supported",
    "n_unsupported",
    "n_uncertain",
    "hallucination_binary_strict",
    "hallucination_binary_lenient",
    "hallucination_count_strict",
    "hallucination_count_lenient",
    "hallucination_rate_strict",
    "hallucination_rate_lenient",
]
for c in hall_numeric_cols:
    if c in hall_df.columns:
        hall_df[c] = pd.to_numeric(hall_df[c], errors="coerce")

# ------------------------------------------------------------------
# Helpers
# ------------------------------------------------------------------
def robust_z(s):
    med = s.median()
    iqr = s.quantile(0.75) - s.quantile(0.25)
    if pd.isna(iqr) or iqr == 0:
        return s - med
    return (s - med) / iqr

def standard_z(a):
    a = np.asarray(a, dtype=float)
    s = a.std(ddof=0)
    if np.isnan(s) or s == 0:
        return a - a.mean()
    return (a - a.mean()) / s

def safe_corr(a, b):
    if len(a) < 2:
        return np.nan
    if np.std(a) == 0 or np.std(b) == 0:
        return np.nan
    return np.corrcoef(a, b)[0, 1]

def get_divergences(idata):
    if "sample_stats" not in idata.groups():
        return np.nan
    if "diverging" not in idata.sample_stats:
        return np.nan
    return int(idata.sample_stats["diverging"].sum().values)

def ppc_interval_check(obs_stat, rep_stats, low=0.05, high=0.95):
    lo = float(np.quantile(rep_stats, low))
    hi = float(np.quantile(rep_stats, high))
    return lo, hi, float(lo <= obs_stat <= hi)

def save_trace_plot(idata, var_names, filename):
    az.plot_trace(idata, var_names=var_names)
    plt.tight_layout()
    plt.savefig(OUT / filename, dpi=150, bbox_inches="tight")
    plt.show()

def save_posterior_plot(idata, var_names, filename):
    az.plot_posterior(idata, var_names=var_names)
    plt.tight_layout()
    plt.savefig(OUT / filename, dpi=150, bbox_inches="tight")
    plt.show()

# ------------------------------------------------------------------
# Build score-model dataframe from feature file
# ------------------------------------------------------------------
feature_cols = ["precision_obs", "recall_obs"]

score_df = (
    df[["case_id"] + feature_cols]
    .replace([np.inf, -np.inf], np.nan)
    .dropna()
    .reset_index(drop=True)
)

X = score_df[feature_cols].copy()
Xr = X.apply(robust_z)

composite_for_orientation = Xr.mean(axis=1).to_numpy()
composite_score = standard_z(composite_for_orientation)

n_cases = len(score_df)
coords = {
    "case": np.arange(n_cases),
}

print("Score model cases:", n_cases)
display(Xr.describe())

# ------------------------------------------------------------------
# Score model
# fixed latent scale, separate residual spreads, one extra loading
# ------------------------------------------------------------------
with pm.Model(coords=coords) as score_model:
    E = pm.Normal("E", mu=0.0, sigma=1.0, dims="case")

    alpha_precision = pm.Normal("alpha_precision", mu=0.0, sigma=0.5)
    alpha_recall = pm.Normal("alpha_recall", mu=0.0, sigma=0.5)
    
    lambda_recall = pm.LogNormal("lambda_recall", mu=0.0, sigma=0.15)
    sigma_precision = pm.HalfNormal("sigma_precision", sigma=0.3)
    sigma_recall = pm.HalfNormal("sigma_recall", sigma=0.3)

    pm.Normal(
        "precision_obs_lik",
        mu=alpha_precision + E,
        sigma=sigma_precision,
        observed=Xr["precision_obs"].values,
        dims="case",
    )

    pm.Normal(
        "recall_obs_lik",
        mu=alpha_recall + lambda_recall * E,
        sigma=sigma_recall,
        observed=Xr["recall_obs"].values,
        dims="case",
    )

    idata_score = pm.sample(
        draws=4000,
        tune=5000,
        chains=4,
        cores=4,
        init="adapt_diag",
        target_accept=0.99,
        max_treedepth=15,
        random_seed=42,
        nuts_sampler="nutpie",
        return_inferencedata=True,
    )

    ppc_score = pm.sample_posterior_predictive(
        idata_score,
        var_names=["precision_obs_lik", "recall_obs_lik"],
        random_seed=42,
    )

idata_score.extend(ppc_score)

# ------------------------------------------------------------------
# Export explainability scores
# ------------------------------------------------------------------
score_summary = az.summary(
    idata_score,
    var_names=[
        "alpha_precision",
        "alpha_recall",
        "lambda_recall",
        "sigma_precision",
        "sigma_recall",
    ],
    round_to=4,
)
score_summary.to_csv(OUT / "score_model_posterior_summary.csv")

score_summary_E = az.summary(
    idata_score,
    var_names=["E"],
    round_to=4,
)
score_summary_E.to_csv(OUT / "score_model_E_summary.csv")

score_mean_raw = idata_score.posterior["E"].mean(dim=("chain", "draw")).values
score_median_raw = idata_score.posterior["E"].median(dim=("chain", "draw")).values

corr = safe_corr(score_mean_raw, composite_for_orientation)
sign = 1.0 if np.isnan(corr) or corr >= 0 else -1.0

scores = score_df[["case_id"]].copy()
scores["explainability_score_mean"] = sign * score_mean_raw
scores["explainability_score_median"] = sign * score_median_raw
scores["composite_score"] = composite_score
scores = scores.sort_values("explainability_score_mean", ascending=False).reset_index(drop=True)
scores.to_csv(OUT / "case_explainability_scores.csv", index=False)

# ------------------------------------------------------------------
# Score diagnostics
# ------------------------------------------------------------------
score_diag_main = az.summary(
    idata_score,
    var_names=[
        "alpha_precision",
        "alpha_recall",
        "lambda_recall",
        "sigma_precision",
        "sigma_recall",
    ],
)
score_diag_E = az.summary(idata_score, var_names=["E"])
score_divergences = get_divergences(idata_score)

save_trace_plot(
    idata_score,
    ["alpha_precision", "alpha_recall", "lambda_recall", "sigma_precision", "sigma_recall"],
    "score_model_trace.png",
)

# ------------------------------------------------------------------
# Score PPC
# ------------------------------------------------------------------
ppc_rows = []
score_ppc_pass_all = True

for feat in feature_cols:
    lik_name = f"{feat}_lik"
    obs = Xr[feat].values
    rep = idata_score.posterior_predictive[lik_name].values.reshape(-1, len(obs))

    rep_means = rep.mean(axis=1)
    rep_sds = rep.std(axis=1, ddof=0)

    obs_mean = float(obs.mean())
    obs_sd = float(obs.std(ddof=0))

    mean_lo, mean_hi, mean_inside = ppc_interval_check(obs_mean, rep_means)
    sd_lo, sd_hi, sd_inside = ppc_interval_check(obs_sd, rep_sds)

    score_ppc_pass_all = score_ppc_pass_all and bool(mean_inside) and bool(sd_inside)

    ppc_rows.append({
        "feature": feat,
        "obs_mean": obs_mean,
        "pp_mean_mean": float(rep_means.mean()),
        "pp_mean_p05": mean_lo,
        "pp_mean_p95": mean_hi,
        "obs_mean_inside_pp_interval": int(mean_inside),
        "obs_sd": obs_sd,
        "pp_sd_mean": float(rep_sds.mean()),
        "pp_sd_p05": sd_lo,
        "pp_sd_p95": sd_hi,
        "obs_sd_inside_pp_interval": int(sd_inside),
    })

ppc_summary = pd.DataFrame(ppc_rows)
ppc_summary.to_csv(OUT / "score_model_ppc_summary.csv", index=False)
display(ppc_summary)

# ------------------------------------------------------------------
# Merge scores with manual hallucination case-level file
# ------------------------------------------------------------------
regdf = hall_df.merge(
    scores[["case_id", "explainability_score_mean", "composite_score"]],
    on="case_id",
    how="inner",
).copy()

regdf = regdf.replace([np.inf, -np.inf], np.nan)
regdf = regdf.dropna(
    subset=[
        "explainability_score_mean",
        "n_reviewed",
        "hallucination_count_strict",
        "hallucination_count_lenient",
        "hallucination_binary_strict",
        "hallucination_binary_lenient",
    ]
).reset_index(drop=True)

regdf["n_reviewed"] = regdf["n_reviewed"].astype(int)
regdf["hallucination_count_strict"] = regdf["hallucination_count_strict"].astype(int)
regdf["hallucination_count_lenient"] = regdf["hallucination_count_lenient"].astype(int)
regdf["hallucination_binary_strict"] = regdf["hallucination_binary_strict"].astype(int)
regdf["hallucination_binary_lenient"] = regdf["hallucination_binary_lenient"].astype(int)

regdf["score_z"] = standard_z(regdf["explainability_score_mean"].to_numpy())
regdf["composite_z"] = standard_z(regdf["composite_score"].to_numpy())

regdf.to_csv(OUT / "manual_hallucination_regression_input.csv", index=False)

print("Regression cases:", len(regdf))
display(regdf.head())

# ------------------------------------------------------------------
# Main model: strict count model
# k unsupported out of n reviewed
# ------------------------------------------------------------------
x = regdf["score_z"].to_numpy()
n_reviewed = regdf["n_reviewed"].to_numpy()
k_strict = regdf["hallucination_count_strict"].to_numpy()
k_lenient = regdf["hallucination_count_lenient"].to_numpy()
y_bin_strict = regdf["hallucination_binary_strict"].to_numpy()
y_bin_lenient = regdf["hallucination_binary_lenient"].to_numpy()

with pm.Model() as strict_count_model:
    beta0 = pm.Normal("beta0", mu=0.0, sigma=1.0)
    beta1 = pm.Normal("beta1", mu=0.0, sigma=1.0)

    p = pm.Deterministic("p", pm.math.sigmoid(beta0 + beta1 * x))

    pm.Binomial(
        "strict_count_obs",
        n=n_reviewed,
        p=p,
        observed=k_strict,
    )

    idata_strict_count = pm.sample(
        draws=3000,
        tune=3000,
        chains=4,
        cores=4,
        init="adapt_diag",
        target_accept=0.99,
        max_treedepth=15,
        random_seed=42,
        nuts_sampler="nutpie",
        return_inferencedata=True,
    )

    ppc_strict_count = pm.sample_posterior_predictive(
        idata_strict_count,
        var_names=["strict_count_obs"],
        random_seed=42,
    )

idata_strict_count.extend(ppc_strict_count)

strict_count_summary = az.summary(
    idata_strict_count,
    var_names=["beta0", "beta1"],
    round_to=4,
)
strict_count_summary.to_csv(OUT / "strict_count_regression_summary.csv")

beta1_samples = idata_strict_count.posterior["beta1"].values.flatten()
prob_beta1_negative = float((beta1_samples < 0).mean())

# ------------------------------------------------------------------
# Robustness 1: lenient count model
# unsupported + uncertain out of n reviewed
# ------------------------------------------------------------------
with pm.Model() as lenient_count_model:
    gamma0 = pm.Normal("gamma0", mu=0.0, sigma=1.0)
    gamma1 = pm.Normal("gamma1", mu=0.0, sigma=1.0)

    p = pm.Deterministic("p", pm.math.sigmoid(gamma0 + gamma1 * x))

    pm.Binomial(
        "lenient_count_obs",
        n=n_reviewed,
        p=p,
        observed=k_lenient,
    )

    idata_lenient_count = pm.sample(
        draws=3000,
        tune=3000,
        chains=4,
        cores=4,
        init="adapt_diag",
        target_accept=0.99,
        max_treedepth=15,
        random_seed=42,
        nuts_sampler="nutpie",
        return_inferencedata=True,
    )

lenient_count_summary = az.summary(
    idata_lenient_count,
    var_names=["gamma0", "gamma1"],
    round_to=4,
)
lenient_count_summary.to_csv(OUT / "lenient_count_regression_summary.csv")

gamma1_samples = idata_lenient_count.posterior["gamma1"].values.flatten()
prob_gamma1_negative = float((gamma1_samples < 0).mean())

# ------------------------------------------------------------------
# Robustness 2: strict binary model
# at least one unsupported hallucination in case
# ------------------------------------------------------------------
with pm.Model() as strict_binary_model:
    delta0 = pm.Normal("delta0", mu=0.0, sigma=1.0)
    delta1 = pm.Normal("delta1", mu=0.0, sigma=1.0)

    p = pm.Deterministic("p", pm.math.sigmoid(delta0 + delta1 * x))

    pm.Bernoulli(
        "strict_binary_obs",
        p=p,
        observed=y_bin_strict,
    )

    idata_strict_binary = pm.sample(
        draws=3000,
        tune=3000,
        chains=4,
        cores=4,
        init="adapt_diag",
        target_accept=0.99,
        max_treedepth=15,
        random_seed=42,
        nuts_sampler="nutpie",
        return_inferencedata=True,
    )

strict_binary_summary = az.summary(
    idata_strict_binary,
    var_names=["delta0", "delta1"],
    round_to=4,
)
strict_binary_summary.to_csv(OUT / "strict_binary_regression_summary.csv")

delta1_samples = idata_strict_binary.posterior["delta1"].values.flatten()
prob_delta1_negative = float((delta1_samples < 0).mean())

# ------------------------------------------------------------------
# Diagnostics and acceptance summary
# ------------------------------------------------------------------
strict_diag = az.summary(idata_strict_count, var_names=["beta0", "beta1"])
lenient_diag = az.summary(idata_lenient_count, var_names=["gamma0", "gamma1"])
binary_diag = az.summary(idata_strict_binary, var_names=["delta0", "delta1"])

strict_div = get_divergences(idata_strict_count)
lenient_div = get_divergences(idata_lenient_count)
binary_div = get_divergences(idata_strict_binary)

save_trace_plot(idata_strict_count, ["beta0", "beta1"], "strict_count_trace.png")
save_posterior_plot(idata_strict_count, ["beta1"], "strict_count_beta1_posterior.png")

acceptance_summary = pd.DataFrame([{
    "score_divergences": int(score_divergences),
    "score_max_rhat_main": float(score_diag_main["r_hat"].max()),
    "score_max_rhat_E": float(score_diag_E["r_hat"].max()),
    "score_ppc_pass": int(score_ppc_pass_all),
    "strict_count_divergences": int(strict_div),
    "strict_count_max_rhat": float(strict_diag["r_hat"].max()),
    "lenient_count_divergences": int(lenient_div),
    "lenient_count_max_rhat": float(lenient_diag["r_hat"].max()),
    "strict_binary_divergences": int(binary_div),
    "strict_binary_max_rhat": float(binary_diag["r_hat"].max()),
    "p_beta1_negative_strict_count": prob_beta1_negative,
    "p_gamma1_negative_lenient_count": prob_gamma1_negative,
    "p_delta1_negative_strict_binary": prob_delta1_negative,
    "overall_minimum_acceptance": int(
        (score_divergences == 0)
        and (float(score_diag_main["r_hat"].max()) <= 1.01)
        and (float(score_diag_E["r_hat"].max()) <= 1.01)
        and score_ppc_pass_all
        and (strict_div == 0)
        and (float(strict_diag["r_hat"].max()) <= 1.01)
    ),
}])

acceptance_summary.to_csv(OUT / "acceptance_summary.csv", index=False)

# ------------------------------------------------------------------
# Final overview
# ------------------------------------------------------------------
final_overview = pd.DataFrame([{
    "n_cases_score_model": len(score_df),
    "n_cases_regression": len(regdf),
    "mean_n_reviewed": float(regdf["n_reviewed"].mean()),
    "mean_strict_rate": float(regdf["hallucination_rate_strict"].mean()),
    "mean_lenient_rate": float(regdf["hallucination_rate_lenient"].mean()),
    "strict_count_beta1_mean": float(beta1_samples.mean()),
    "strict_count_beta1_median": float(np.median(beta1_samples)),
    "strict_count_beta1_q05": float(np.quantile(beta1_samples, 0.05)),
    "strict_count_beta1_q95": float(np.quantile(beta1_samples, 0.95)),
    "strict_count_p_beta1_negative": prob_beta1_negative,
    "lenient_count_gamma1_mean": float(gamma1_samples.mean()),
    "lenient_count_p_gamma1_negative": prob_gamma1_negative,
    "strict_binary_delta1_mean": float(delta1_samples.mean()),
    "strict_binary_p_delta1_negative": prob_delta1_negative,
    "overall_minimum_acceptance": int(acceptance_summary["overall_minimum_acceptance"].iloc[0]),
}])
final_overview.to_csv(OUT / "final_overview.csv", index=False)

display(score_summary)
display(scores.head(10))
display(strict_count_summary)
display(lenient_count_summary)
display(strict_binary_summary)
display(acceptance_summary)
display(final_overview)

print("\\nSaved outputs to:", OUT.resolve())
```
